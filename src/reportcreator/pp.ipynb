```python
import json
from pathlib import Path

PROJECT_ROOT = Path("C:/Users/kelly/working/llm_report_project")
with open(PROJECT_ROOT / "data/chunks/chunking_03_hybrid.json", encoding='utf-8') as f:
    data = json.load(f)

chunks = data['chunks']
# 유안타증권 기타 섹터에 배터리 관련 청크가 있는지 확인
result = [
    c for c in chunks 
    if c.get('source_firm') == '유안타증권'
    and c.get('sector') == '기타'
    and any(kw in c.get('title', '') + c.get('page_content', '')[:200] 
            for kw in ['배터리', '2차전지', 'ESS', '전기차'])
]
print(f'유안타증권 배터리 관련 청크: {len(result)}')
for c in result[:5]:
    print(f"  제목: {c.get('title','')[:40]}")
```
